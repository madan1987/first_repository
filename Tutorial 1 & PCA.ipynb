```python
import numpy as np
from numpy import linalg
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn import preprocessing
```

Data

```python
x = np.dot(np.random.random(400).reshape(200,2),np.random.random(4).reshape(2,2))
```

```python
x.shape
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
x= preprocessing.scale(x)
plt.scatter(x[:,0],x[:,1])
plt.grid()
```

PCA

```python
pca = PCA(n_components=2)
pca.fit(x)
```

```python
print(pca.components_)
```

```python
print(pca.explained_variance_)
```

Linalg Eig

```python
covar = np.cov(x.T)
w, v = linalg.eig(covar)
```

```python
w
```

```python
v
```

Linalg SVD

```python
u, s, vh = np.linalg.svd(covar)
```

```python
u
```

```python
s
```

```python
vh
```

PLOT

```python
plt.scatter(x[:,0],x[:,1],alpha=0.5)

plt.quiver(0,0,u[:,0],u[:,1],scale=3)

plt.quiver(0,0,v.T[:,0],v.T[:,1],scale=4,color='red')

plt.quiver(0,0,pca.components_[:,0],pca.components_[:,1],scale=3,color='blue')


plt.grid()
```
